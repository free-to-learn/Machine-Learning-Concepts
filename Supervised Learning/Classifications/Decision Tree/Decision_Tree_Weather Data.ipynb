```python
# Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()  # Conveating the Non Numerical Data into Numerical
```

```python
# Converting string labels into numbers.
Data['Outlook']=le.fit_transform(Data['Outlook'])
Data.head()
```

```python
# Converting string labels into numbers.
Data['Temp']=le.fit_transform(Data['Temp'])
Data['Humidity'] = le.fit_transform(Data['Humidity'])
Data['Windy']= le.fit_transform(Data['Windy'])
Data['Play']= le.fit_transform(Data['Play'])
Data.head()
```

```python
Feature = Data.drop(['Play'],axis=1)
Feature.head()
```

```python
Label = Data['Play']
Label[0:5]
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(Feature, Label)
clf
```

```python
clf.predict([[0,1,0,0]]) # yes i can play the match
```

```python
clf.classes_ # Yes or No
```

```python
clf.n_classes_
```

```python
clf.n_features_ # outlook, Temp,Humidity,windy
```

```python
clf.score(Feature,Label)
```
